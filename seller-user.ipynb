```python
dfu28456538["user count"] = dfu28456538.groupby("user id")["user id"].transform('count')
dfu28456538.sort_values('user count', ascending=False)

dfu28456538["seller count"] = dfu28456538.groupby("Seller Id")["Seller Id"].transform('count')
dfu28456538.sort_values('seller count', ascending=False)
```

```python
dfu28456538 = dfu28456538[['user id', 'seller count']].groupby('Seller Id').agg({'user id': 'max', 'seller count': 'sum'}).sort_values('seller count', ascending=False)
```
