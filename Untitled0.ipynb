```python
def get_len_unique_users (si):

  temp = df.loc[(df['Seller Id'] == si)]
  y = len(set(temp['user id'].values))

  return(y)

get_len_unique_users(91055)  
```

```python
tsdf['Total unique users'] = tsdf['Seller Id'].apply(lambda x: get_len_unique_users(x))
tsdf
```
